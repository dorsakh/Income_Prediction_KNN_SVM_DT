```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score
model_KNN = KNeighborsClassifier(leaf_size=5 ,n_neighbors=15 , algorithm='auto')
model_KNN.fit(x_train , y_train)
# model.score(x_train,y_train)
yKNN_predict = model_KNN.predict(x_test)
accuracy_score(y_test, yKNN_predict)
```

The k-nearest neighbors algorithm, also known as KNN or k-NN, is a non-parametric,<br>
supervised learning classifier, which uses proximity to make classifications or predictions about the grouping of an individual data point<br>
We use KNeighborsClassifier to build a model with proper attributes.<br>
5 is chosen for leaf_size because it gives us the best accuracy with cross validation(k=12)<br>
The n_neighbors with the value of 15 gives th most proper accuracy for both train and test sets<br>
Algorithm "auto" will attempt to decide the most appropriate algorithm based on the values passed to fit method.<br>


```python
model_KNN.score(x_train , y_train)
```

estimate model accuracy  for train_set

```python
n = np.arange(1,21)

atrain = np.empty(20)   
atest  = np.empty(20)

for i,k in enumerate(n):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    atrain[i]= model.score(x_train,y_train)
    atest[i] = model.score(x_test,y_test)
    
plt.plot(n , atrain , label = 'Traing Accuracy')
plt.plot(n , atest  , label = 'Testing Accuracy')   
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
```

F1 = 2 * (precision * recall) / (precision + recall)
